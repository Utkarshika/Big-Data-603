```python
#importing the necessary libraries

import time 
import json 
from kafka import KafkaProducer, KafkaConsumer
import matplotlib
import requests
```

```python
from time import sleep

# Messages will be serialized as JSON 
def message_serializer(message):
    return json.dumps(message).encode('utf-8')

# Kafka Producer
producer = KafkaProducer(
    bootstrap_servers=['localhost:9092'],
    value_serializer= message_serializer
)


if __name__ == '__main__':
    t_end = time.time() + 60 * 60
    #here the loop runs for 1 hour
    while time.time() < t_end:
        url="http://api.open-notify.org/iss-now.json"
        response = urlopen(url)
        json_data = json.loads(response.read())
        print(json_data)
        sent_message = json_data
        #sedning the message in json format to consumer
        producer.send('ISS_data', json_data)
        # message sent after a delay of 5 seconds
        time.sleep(5)

```
